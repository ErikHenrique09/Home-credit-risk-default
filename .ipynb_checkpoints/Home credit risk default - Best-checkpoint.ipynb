<center><h1>Home Credit Risk Default</h1></center>

Muitas pessoas lutam para obter empréstimos devido a históricos de crédito insuficientes ou inexistentes. E, infelizmente, essa população é muitas vezes aproveitada por credores não confiáveis.

### Grupo de crédito residencial

O Crédito Habitação busca ampliar a inclusão financeira da população não bancarizada, proporcionando uma experiência de empréstimo positiva e segura. Para garantir que essa população carente tenha uma experiência de empréstimo positiva, o Home Credit utiliza uma variedade de dados alternativos - incluindo informações de telecomunicações e transacionais - para prever a capacidade de pagamento de seus clientes.

### Objetivo geral

Embora o Home Credit esteja atualmente usando vários métodos estatísticos e de aprendizado de máquina para fazer essas previsões, eles estão desafiando os Kagglers a ajudá-los a liberar todo o potencial de seus dados. Isso garantirá que os clientes com capacidade de pagamento não sejam rejeitados e que os empréstimos sejam concedidos com um calendário de principal, vencimento e reembolso que capacitará seus clientes a serem bem-sucedidos.

# Treinando modelo com os dados de treino
Teste de varios modelos na base sem alterações e normalizada, para termos opções

```python
x = df.drop('TARGET', axis=1)
x_norm = df_normaliza.drop('TARGET', axis=1)

y = df['TARGET']
y_norm = df_normaliza['TARGET']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3)
#xtrain, xtest, ytrain, ytest = train_test_split(x_norm,y_norm,test_size=0.3)
```

```python
xtrain.shape, xtest.shape
```

```python
# Modelos
model_rfc = RandomForestClassifier(n_estimators = 100, random_state = 50, n_jobs = -1) 
model_gbx = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0) 
model_dtc = AdaBoostClassifier(n_estimators=100, random_state=0) 
model_lgbm = LGBMClassifier() 
model_xgb = XGBClassifier(n_estimators = 100, max_depth = 10, learning_rate = 0.01, subsample = 1,  random_state=0)
```

```python
model_rfc.fit(xtrain,ytrain)
model_gbx.fit(xtrain,ytrain)
model_dtc.fit(xtrain,ytrain)
model_xgb.fit(xtrain, ytrain)
model_lgbm.fit(xtrain,ytrain)
```

```python
# Probabilidades de ser 1 (% de risco)
predc1_proba = model_rfc.predict_proba(xtest)[0:][0:,1]
predc2_proba = model_gbx.predict_proba(xtest)[0:][0:,1]
predc3_proba = model_dtc.predict_proba(xtest)[0:][0:,1]
predc4_proba = model_lgbm.predict_proba(xtest)[0:][0:,1]
predc5_proba = model_xgb.predict_proba(xtest)[0:][0:,1]
```

```python
# Resultado 
predc1 = model_rfc.predict(xtest)
predc2 = model_gbx.predict(xtest)
predc3 = model_dtc.predict(xtest)
predc4 = model_lgbm.predict(xtest)
predc5 = model_xgb.predict(xtest)
```

## Avaliando na base de treino

```python
print(f"Score model_rfc:{ model_rfc.score(xtest, ytest) }")
print(f"Score model_gbx:{ model_gbx.score(xtest, ytest)}")
print(f"Score model_dtc:{ model_dtc.score(xtest, ytest)}")
print(f"Score model_lgbm:{ model_lgbm.score(xtest, ytest)}")
print(f"Score model_xgb:{ model_xgb.score(xtest, ytest)}")
```

- O score do modelo não apresentou ser uma boa metrica, ja que todos se sairam semelhantes porem o modelo LGBMClassifier foi melhor em todos os casos

```python
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(ytest, predc1_proba))}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(ytest, predc2_proba))}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(ytest, predc3_proba))}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(ytest, predc4_proba))}")
print(f"RMSE: {np.sqrt(metrics.mean_squared_error(ytest, predc5_proba))}")
```

- O RMSE acabou sendo mais explicativo, ja que o menor RMSE foi do modelo que melhor perfomou nas analises

```python
print(f'{"-="*28}\n\tModelo RandomForesClassifier:\n{"-="*28}\n{classification_report(ytest, predc1)}', end= '')
print(f'{"-="*28}\n\tModelo GradientBoostingClassifier:\n{"-="*28}\n{classification_report(ytest, predc2)}')
print(f'{"-="*28}\n\tModelo AdaBoostClassifier:\n{"-="*28}\n{classification_report(ytest, predc3)}')
print(f'{"-="*28}\n\tModelo LGBMClassifier:\n{"-="*28}\n{classification_report(ytest, predc4)}')
print(f'{"-="*28}\n\tModelo XGBoost:\n{"-="*28}\n{classification_report(ytest, predc5)}')
```

- Essa tabela foi analisada com enfase na precisão de prever inadimplentes(1) ja que os lucros com um cliente não inadimplente podem não cobrir um emprestimo sem retorno, e mesmo a RandomForest tendo uma maior precisão na base de treino, ao ser utilizada para responder o kaggle o LGBMClassifier se saiu melhor na maioria das vezes

```python
pdcs = [predc1_proba,predc2_proba,predc3_proba,predc4_proba, predc5_proba]
modelss = ['RandomForesClassifier' ,'GradientBoostingClassifier','AdaBoostClassifier' ,'LGBMClassifier' ,'XGBoost']

fig, ax = plt.subplots(1, 5)
axs = list(ax.flat)

for i, ax in enumerate(axs):
    rfp, rvp,lim = roc_curve(ytest, pdcs[i])
    # Gráfico da curva roc
    ax.set_title(modelss[i])
    ax.set_xlabel('1- Especificidade')
    ax.set_ylabel('Sensibilidade')
    ax.plot(rfp, rvp, marker='.', label='KNN',color="orange", )
    ax.plot([0, 1], [0, 1], color='darkblue', linestyle='--')

    
plt.subplots_adjust(left=0.1, bottom=2, right=3, top=3, wspace=0.3, hspace=34)
```

- A maior area sobre a curva roc é do modelo LGBMClassifier em quase todos os testes
- Os modelos alcançam uma pontuação maior com os dados sem nenhuma padronização/normalização, e ate mesmo sem o tratamento de outliers.
- Por ser um problema de classificação, alguns outliers podem ser dados fortemente explicativos e indicar diretamente um alto ou baixo risco de inadimplencia
